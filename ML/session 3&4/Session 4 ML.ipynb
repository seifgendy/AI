```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.linear_model import LinearRegression , Lasso , Ridge
```

```python
x = np.linspace(1 , 20 , 50).reshape(-1,1)
y = np.linspace(10000 , 50000 , 50)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x , y )
```

```python
lr = LinearRegression().fit(x,y)
```

```python
y_pred_lr = lr.predict(x)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x , y )
plt.plot(x,y_pred_lr , color = "red")
```

```python
ls = Lasso(alpha= 10000).fit(x,y)
```

```python
y_pred_ls = ls.predict(x)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x , y )
plt.plot(x,y_pred_lr , color = "red")
plt.plot(x,y_pred_ls , color = "black")
```

```python
rd = Ridge(alpha=10000).fit(x,y)
```

```python
y_pred_rd = rd.predict(x)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x , y )
plt.plot(x,y_pred_lr , color = "red")
plt.plot(x,y_pred_ls , color = "black")
plt.plot(x,y_pred_rd , color = "yellow")
```
