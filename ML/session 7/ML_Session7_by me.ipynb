```python
new_df = df.drop(columns = ["Unnamed: 32" , "id"])
```

```python
#we can use label encoder or just a replace to turn our target from letters to number
df["diagnosis"].value_counts()
```

```python
le = LabelEncoder()
new_df["diagnosis"] = le.fit_transform(df["diagnosis"])
```

```python
le.classes_
```

```python
#we do this so that i take the out put prediction in a meaningfull form
le.inverse_transform([0,1])
```

```python
new_df.head()
```

```python
x =  new_df.drop(columns = ["diagnosis"] , axis=1)
y = new_df["diagnosis"]
```

```python
#stratify make it choose rows in a fair way between the terget so that not train much on only one target
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.75 , random_state = 42
                                                       ,stratify = y)
```

```python
model = LogisticRegression(max_iter=10000)
#increased number of iterations so that the model can converge or i could have used scaling to standardize the data
```

```python
model.fit(x_train , y_train)
```

```python
y_pred_train = model.predict(x_train)
```

```python
cm = confusion_matrix(y_train , y_pred_train)
cm
```

```python
df_cm = pd.DataFrame(cm , columns =["positive_pred" , "negative_pred"]
             , index = ["positive_actual" , "negative_actual"])
df_cm
```

```python
print(classification_report(y_train , y_pred_train))
# 89 = 86 + 3
# 53 = 47 + 6
```

since the accuracies are near 96 & 94 then no overfit and if both were low then underfit


```python
#real model test since it is done on test data not train data
y_pred = model.predict(x_test)
print(classification_report(y_test , y_pred))
```
