```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , classification_report 
```

```python
df = pd.read_csv("Breast_Cancer.csv")
```

```python
df.isna().sum()
```

```python
new_df = df.drop(columns=["Unnamed: 32" , "id"])
```

```python
df["diagnosis"].value_counts()
```

```python
le = LabelEncoder()
new_df["diagnosis"] = le.fit_transform(df["diagnosis"])
```

```python
le.classes_
```

```python
le.inverse_transform([0,1])
```

```python
new_df.head(3)
```

```python
x = new_df.drop("diagnosis" ,axis=1)
y = new_df["diagnosis"]
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x , y , train_size=.75, 
                                                      random_state=42,stratify=y) 
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train , y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
confusion_matrix(y_test , y_pred)
```

```python
# Change figure size and increase dpi for better resolution
plt.figure(figsize=(5,4), dpi=100)
# Scale up the size of all text
sns.set(font_scale = 1.1)
 
ax = sns.heatmap(confusion_matrix(y_test , y_pred), annot=True, fmt='d', )
 
# set x-axis label and ticks. 
ax.set_xlabel("Predicted class", fontsize=14, labelpad=20)
ax.xaxis.set_ticklabels(['B', 'M'])
 
# set y-axis label and ticks
ax.set_ylabel("Actual class", fontsize=14, labelpad=20)
ax.yaxis.set_ticklabels(['B', 'M'])
 
# set plot title
ax.set_title("Confusion Matrix", fontsize=14, pad=20)
 
plt.show()
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train_scale , y_train)
```

```python
y_pred = knn.predict(x_test_scale)
```

```python
confusion_matrix(y_test , y_pred)
```

```python
# Change figure size and increase dpi for better resolution
plt.figure(figsize=(5,4), dpi=100)
# Scale up the size of all text
sns.set(font_scale = 1.1)
 
ax = sns.heatmap(confusion_matrix(y_test , y_pred), annot=True, fmt='d', )
 
# set x-axis label and ticks. 
ax.set_xlabel("Predicted class", fontsize=14, labelpad=20)
ax.xaxis.set_ticklabels(['B', 'M'])
 
# set y-axis label and ticks
ax.set_ylabel("Actual class", fontsize=14, labelpad=20)
ax.yaxis.set_ticklabels(['B', 'M'])
 
# set plot title
ax.set_title("Confusion Matrix", fontsize=14, pad=20)
 
plt.show()
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsClassifier()
param = {
    "n_neighbors": range(1,11)
}
grid_search = GridSearchCV(estimator=knn , cv = 5 , param_grid = param , 
                           scoring= "accuracy"  , n_jobs=-1 , verbose=3)
grid_search.fit(x_train_scale , y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(x_test_scale)
```

```python
confusion_matrix(y_test , y_pred)
```

```python
# Change figure size and increase dpi for better resolution
plt.figure(figsize=(5,4), dpi=100)
# Scale up the size of all text
sns.set(font_scale = 1.1)
 
ax = sns.heatmap(confusion_matrix(y_test , y_pred), annot=True, fmt='d', )
 
# set x-axis label and ticks. 
ax.set_xlabel("Predicted class", fontsize=14, labelpad=20)
ax.xaxis.set_ticklabels(['B', 'M'])
 
# set y-axis label and ticks
ax.set_ylabel("Actual class", fontsize=14, labelpad=20)
ax.yaxis.set_ticklabels(['B', 'M'])
 
# set plot title
ax.set_title("Confusion Matrix", fontsize=14, pad=20)
 
plt.show()
```
