```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier , StackingClassifier
from sklearn.ensemble import BaggingClassifier , VotingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
dataset = load_iris()
```

```python
X = dataset.data
Y = dataset.target
```

# VotingClassifier

```python
base_model = [
    ('lr' , LogisticRegression(max_iter=1000)),
    ('dt' , DecisionTreeClassifier(random_state=42)),
    ('rf' , RandomForestClassifier(n_estimators=100)),
    ('knn', KNeighborsClassifier(n_neighbors=5)),
    ('svm', SVC())
]
```

```python
voting_clf = VotingClassifier(estimators=base_model , voting="hard" ) # voting = "soft"
```

```python
voting_clf.fit(X , Y )
```

```python
y_pred_vot = voting_clf.predict(X)
```

```python
accuracy_score(Y , y_pred_vot)
```

# StackingClassifier

```python
base_model = [
    ('lr' , LogisticRegression(max_iter=1000)),
    ('dt' , DecisionTreeClassifier(random_state=42)),
    ('rf' , RandomForestClassifier(n_estimators=100)),
    ('knn', KNeighborsClassifier(n_neighbors=5)),
    ('svm', SVC())
]
```

```python
meta_model = LogisticRegression(max_iter=1000)
```

```python
stacking_clf = StackingClassifier(estimators=base_model , final_estimator= meta_model , cv = 5 )
```

```python
stacking_clf.fit(X , Y)
```

```python
y_pred_stack = stacking_clf.predict(X)
```

```python
accuracy_score(Y , y_pred_stack)
```

# BaggingClassifier

```python
bagging = BaggingClassifier(estimator= DecisionTreeClassifier() ,n_estimators= 150 , random_state=42 )
```

```python
bagging.fit(X , Y )
```

```python
y_pred_bagging = bagging.predict(X)
```

```python
accuracy_score(Y , y_pred_bagging)
```

# GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
boosting_clf = GradientBoostingClassifier(n_estimators= 50 , random_state=42)
```

```python
boosting_clf.fit(X , Y)
```

```python
y_pred_boosting = boosting_clf.predict(X)
```

```python
accuracy_score(Y , y_pred_boosting)
```
