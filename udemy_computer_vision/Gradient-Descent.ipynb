## Iterate Hot and Cold Learning Method
- weight = 0.5
- input_score = 0.5
- ground_truth_prediction = 0.08
- learning_rate = 0.001
- iteration in range(331)

**Neural Network Prediction: 0.08499999999999985**

```
weight = 0.5
input_score = 0.5
ground_truth_prediction = 0.08
learning_rate = 0.001

def simple_neural_network(input, weight):
    neural_network_prediction = input * weight
    return neural_network_prediction

for iteration in range(331):

    neural_network_prediction = simple_neural_network(input_score, weight)
    square_error = (neural_network_prediction - ground_truth_prediction ) ** 2

    print("Square Error:" + str(square_error) + " Neural Network Prediction: " + str(neural_network_prediction))
    
    prediction_increasing_weight = simple_neural_network(input_score, (weight + learning_rate))
    square_error_increasing_weight = (ground_truth_prediction - prediction_increasing_weight) ** 2
    print("Prediction while Increasing Weight:" + str(prediction_increasing_weight))
    
    prediction_decreasing_weight = simple_neural_network(input_score, (weight - learning_rate))
    square_error_decreasing_weight = (ground_truth_prediction - prediction_decreasing_weight) ** 2
    print("Prediction While Decreasing Weight:" + str(prediction_decreasing_weight))
    print("\n")
    
    if(prediction_decreasing_weight < prediction_increasing_weight):
        weight = weight - learning_rate
    else:
        weight = weight + learning_rate
```

## Gradient Descent
- weight = 0.3
- input_score = 0.3
- ground_truth_prediction = 0.4
- iteration in range(40)

## Calculation of Gradient Descent
- square_error = (neural_network_prediction - ground_truth_prediction ) ** 2
- direction_and_amount = (neural_network_prediction - ground_truth_prediction ) * input
- weight = weight - direction_and_amount

**Prediction: 0.3921661366013863**

```
weight = 0.3
input_score = 0.3
ground_truth_prediction = 0.4

for iteration in range(40):
    neural_network_prediction = input_score * weight
    square_error = (neural_network_prediction - ground_truth_prediction ) ** 2

    direction_and_amount = (neural_network_prediction - ground_truth_prediction ) * input_score
    weight = weight - direction_and_amount

    print("Square Error:" + str(square_error) + " Prediction: " + str(neural_network_prediction))
```

## Overcoming The Limitation of Gradient Descent
- weight = 0.3
- input_score = 2
- ground_truth_prediction = 0.4
- iteration in range(40)
- alpha = 0.1

## Calculation of Gradient Descent
- square_error = (neural_network_prediction - ground_truth_prediction ) ** 2
- direction_and_amount = (neural_network_prediction - ground_truth_prediction ) * input
- weight = weight - (alpha * direction_and_amount)

```


weight = 0.3
input_score = 2
ground_truth_prediction = 0.4
alpha = 0.1

for iteration in range(40):
    neural_network_prediction = input_score * weight
    square_error = (neural_network_prediction - ground_truth_prediction ) ** 2

    direction_and_amount = (neural_network_prediction - ground_truth_prediction ) * input_score
    weight = weight - (alpha * direction_and_amount)

    print("Error:" + str(square_error) + " Prediction: " + str(neural_network_prediction))
```
